```python
data2=data.iloc[:,:9]
data2
```

```python
data2.isnull().sum()
```

```python
data2.loc[data['GRE Score'].isnull()==True,'GRE Score']=np.median(data2['GRE Score'].dropna(axis=0))
```

```python
data2.loc[data['TOEFL Score'].isnull()==True,'TOEFL Score']=np.median(data2['TOEFL Score'].dropna(axis=0))
```

```python
data2.loc[data['University Rating'].isnull()==True,'University Rating']=np.median(data2['University Rating'].dropna(axis=0))
```

```python
data2.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
data2[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']]=scale.fit_transform(data[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']])
```

## Model Creation

```python
#Step1: Independent and dependent variable creation
x=data2[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research']]
y=data2['Chance of Admit']
```

```python
# step2: Training and testing data creation
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
x_test
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
##Step 3: Creating Model
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
y_predict=LR.predict(x_test)
```

```python
##Step4: Model evaluation
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_predict)
print('r2score of model: ',r2score)
```
